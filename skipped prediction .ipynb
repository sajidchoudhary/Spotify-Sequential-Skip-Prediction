# Feature Selection

```python
# check correlation between continous independent features.
plt.figure(figsize=(20,15))
sns.heatmap(continuous_features.corr(),annot=True,cmap="seismic_r")
plt.show()
```

```python
corr_col=[]
def correlation (data_frame,threshold):
    corr_matrix = data_frame.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>=threshold:
                col_name = corr_matrix.columns[i]
                corr_col.append(col_name)
```

```python
# 90% or more than 90% correlated features.
correlation (continuous_features,0.9)
corr_col
```

```python
# drop bounciness and dyn_range_mean features

x_df = x_df.drop(['bounciness', 'dyn_range_mean'],axis=1)
x_df
```

```python
from sklearn.feature_selection import f_classif
```

```python
f_test = f_classif(x_df,y_df)
f_test = pd.DataFrame(f_test)
f_test = f_test.T
f_test.columns = ["f_score","p-value"]
f_test.index = x_df.columns
p_value = f_test.sort_values("p-value")
less_p_value = p_value[p_value['p-value']<=0.05]
less_p_value
```

```python
x_df = x_df.drop(less_p_value.index,axis=1)
x_df
```

```python
# train and validation data split.

from sklearn.model_selection import train_test_split

x_train,x_valid,y_train,y_valid = train_test_split(x_df,y_df,test_size=0.25,random_state=123)
```
